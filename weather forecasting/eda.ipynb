Cho đến năm 2015, tốc độ gió cao hơn trong mùa gió mùa (tháng 8 và tháng 9) và mùa gió mùa rút lui (tháng 12 và tháng 1) .Sau năm 2015, không có sự bất thường nào về tốc độ gió trong mùa gió mùa. Bây giờ chúng ta hãy xem xét mối quan hệ giữa nhiệt độ và độ ẩm:


```python
figure = px.scatter(data_frame=data, x="humidity", y="meantemp", size="meantemp", trendline="ols",
                    title="Relationship Between Temperature and Humidity")
figure.show()
```

Có mối tương quan âm giữa nhiệt độ và độ ẩm ở Delhi. Điều này có nghĩa là nhiệt độ cao hơn dẫn đến độ ẩm thấp và nhiệt độ thấp hơn dẫn đến độ ẩm cao.


Phân tích sự thay đổi nhiệt độ

```python
data["date"] = pd.to_datetime(data["date"], format = '%Y-%m-%d')
data['year'] = data['date'].dt.year
data["month"] = data["date"].dt.month
```

```python
data.head()
```

Xem xét sự thay đổi nhiệt độ ở Delhi qua các năm:

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```

Forecasting Weather

The Facebook prophet model is one of the best techniques for time series forecasting.

```python
forecast_data = data.rename(columns = {"date": "ds", 
                                       "meantemp": "y"})
```

```python
forecast_data
```

```python
from prophet import Prophet 
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
